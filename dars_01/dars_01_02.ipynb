```python
# kutubxalarni chaqirish

import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# datani chaqirib olish

df = pd.read_csv('house_price.txt')
```

```python
df.info()
```

```python
# tushurib qoldirilgan qiymatlarni tekshirish

df.isnull().sum()
```

```python
# tushurib qoldirilgan qiymatlarni mode yoli mean orqali to'ldirish

def tozala(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=='object':
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(),inplace=True)
    return df
```

```python
tozala(df)
```

```python
df.info()
```

```python
# tushurib qoldiriligan qiymatlar toldirildi

df.isnull().sum()
```

```python
# Matn ko'rinishdan raqamli korinishga almashtrish kutubxonasi

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# Matn ko'rinishdan raqamli korinishga almashtrish
def encodlash(df):
    for col in df.columns:
        if df[col].dtype == 'object':
            if df[col].nunique() <= 6:
                dummies = pd.get_dummies(df[col], prefix=col, dtype=int)
                df = pd.concat([df.drop(columns=[col]), dummies], axis=1)
            else:
                df[col] = le.fit_transform(df[col])
    
    return df
```

```python
df=encodlash(df)
```

```python
# katta qiymatdagi sonlarni xil xil korinishga kelgirish , yani 0 va 1 oralig'iga keltirish kutubxonasi

from sklearn.preprocessing import StandardScaler
scaling = StandardScaler()
```

```python
# katta qiymatdagi sonlarni xil xil korinishga kelgirish , yani 0 va 1 oralig'iga keltirish

def scaling_qil(df):
    numerical_col = df.select_dtypes(include=['int32','int64','float64']).columns.drop('totalPrice')
    df[numerical_col]=scaling.fit_transform(df[numerical_col])
    return df
```

```python
df=scaling_qil(df)
```

```python
df.head()
```

```python
df.info()
```

```python
# datalarni x va y qiymatlarga bolish

x = df.drop(['totalPrice'], axis=1)
y = df[['totalPrice']]
```

```python
# datani train qilish

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# kerakli algorithmni tanlab uni chaqirib olish

from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# modelga orgatish
model.fit(x_train, y_train)
```

```python
# orgatilgan modelni predict qilish

model_pred=model.predict(x_test)
```

```python
model_pred
```

```python
# biz modelinmizni qanchalik yaxshi ishlayotgan ekanligini tekshirishimiz kerak 

from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("R² (Tushuntirilgan dispersiya):", r2_score(y_test, model_pred))
print("MSE (Xatolik):", mean_squared_error(y_test, model_pred))
```
